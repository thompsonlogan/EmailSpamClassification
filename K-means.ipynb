```python
# Imports
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.cluster import KMeans

from sklearn import metrics

import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('emails.csv')
feature_names = data.columns
# print("Features", feature_names)
y = data.iloc[:, -1] # last column the predicition
x = data.iloc[:, 1:-1] # all columns except the first and last

cm_labels = ['spam', 'not spam']
```

```python
#Scaling
x = (x - np.min(x)) / (np.max(x) - np.min(x)) #Normalization
```

```python
# Train the model
kmeans = KMeans(n_clusters=2)
y_km = kmeans.fit_predict(x)
```

```python
# Graph the Data
x = np.array(x)

# plot the 2 clusters
plt.scatter(
    x[y_km == 0, 0], x[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    x[y_km == 1, 0], x[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

# plot the centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()

print('cluster 1 size: ' + str(x[y_km == 0,0].size))
print('cluster 2 size: ' + str(x[y_km == 1,0].size))
print('Benign - count: ' + str(np.sum(y == 1)))
print('Malignant - count: ' + str(np.sum(y == 0)))
```
